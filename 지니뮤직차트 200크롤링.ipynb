```python
from bs4 import BeautifulSoup
from selenium import webdriver

driver = webdriver.Chrome()
data = []
page_num = 1
songs_per_page = 50

while True:
    driver.get(f'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230601&hh=17&rtm=Y&pg={page_num}')
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    titles = soup.findAll('a', {'class':'title ellipsis'})
    artists = soup.findAll('a', {'class':'artist ellipsis'})

    for i, (t, a) in enumerate(zip(titles,artists)):
        title = t.text.strip()
        artist = a.text.strip()
        rank = (page_num - 1) * songs_per_page + (i + 1)
        data.append([rank, artist, title])
        print('{0:3d}위 {1} - {2}'.format(rank, artist, title))

    page_num += 1

   
    if page_num > 4:
        break

driver.quit()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data, columns=['Rank', 'Artist', 'Title'])
print(df)
```
